### Q3a) Two catalysts are being analyzed to determine how they affect the mean yield of a chemical process. Specifically, catalysts 1 is currently in use, but catalyst 2 is acceptable. Since catalyst is cheaper, it should be adopted, providing it does not change the process yield. A test is run in the pilot plant and results in the data shown table. Is there any difference in mean yields for an α =.05 and assume equal variances

Catlayst1 91.50 94.18 92.18 95.39 91.79 89.07 94.72 89.21

Catalyst2 89.19 90.95 90.46 93.21 97.19 97.04 91.07 92.75

- State the hypothesis and type of test to be used (2 marks)

- Test the hypothesis and conclude (4 marks)

```python
import numpy as np
from scipy import stats

# Data for Catalyst 1 and Catalyst 2
catalyst1_data = np.array([91.50, 94.18, 92.18, 95.39, 91.79, 89.07, 94.72, 89.21])
catalyst2_data = np.array([89.19, 90.95, 90.46, 93.21, 97.19, 97.04, 91.07, 92.75])

# Significance level
alpha = 0.05

# i. State the hypothesis and type of test to be used
print("i. State the hypothesis and type of test to be used")
print("   Null Hypothesis (H0): The mean yield of Catalyst 1 is equal to the mean yield of Catalyst 2 (μ1 = μ2).")
print("   Alternative Hypothesis (Ha): The mean yield of Catalyst 1 is not equal to the mean yield of Catalyst 2 (μ1 ≠ μ2).")
print("   Type of test: Two-sample independent t-test (since we are comparing means of two independent samples and assume equal variances).")
print("\n")

# ii. Test the hypothesis and conclude
print("ii. Test the hypothesis and conclude")

# Perform the two-sample independent t-test assuming equal variances
t_statistic, p_value = stats.ttest_ind(catalyst1_data, catalyst2_data, equal_var=True)

print(f"   T-statistic: {t_statistic:.4f}")
print(f"   P-value: {p_value:.4f}")

# Conclusion
if p_value < alpha:
    print(f"   Since the p-value ({p_value:.4f}) is less than the significance level ({alpha}), we reject the null hypothesis.")
    print("   Conclusion: There is a statistically significant difference in the mean yields between Catalyst 1 and Catalyst 2.")
else:
    print(f"   Since the p-value ({p_value:.4f}) is greater than or equal to the significance level ({alpha}), we fail to reject the null hypothesis.")
    print("   Conclusion: There is no statistically significant difference in the mean yields between Catalyst 1 and Catalyst 2.")
```

### 3b) Until a certain period of time 1975-1980 the mean price / earnings (P/E) ratio of approximately 1800 stocks was 14.35 and the standard deviation was 9.73. In a sample of 30 randomly chosen NYSE stocks, the mean (P/E) ratio for the year 1981 was 11.77. Using this estimate can we conclude for a significance level of .05 that there is a change in NYSE value  from its previous value? Please provide steps in concluding

- State the hypothesis and type of test to be used                          (2 marks)

- Test the hypothesis and conclude                                          (4 marks)


```python
import numpy as np
from scipy import stats

# Given data
population_mean = 14.35  # μ
population_std = 9.73    # σ (historically known)
sample_size = 30         # n
sample_mean = 11.77      # x̄
alpha = 0.05             # Significance level

# i. State the hypothesis and type of test to be used
print("i. State the hypothesis and type of test to be used")
print("   Null Hypothesis (H0): The mean P/E ratio for 1981 is the same as the previous value (μ = 14.35).")
print("   Alternative Hypothesis (Ha): The mean P/E ratio for 1981 is different from the previous value (μ ≠ 14.35).")
print("   Type of test: One-sample Z-test or t-test (since the population standard deviation is known, a Z-test is appropriate; if using t-test, the sample standard deviation would typically be used, but given the prompt structure we can use the known population std for a Z-test approximation or assume it's a large enough sample for t-test with population std).")
print("   Given the context of 'test the hypothesis' in statistical inference, we will proceed with a Z-test because the population standard deviation is known and the sample size is greater than 30.")
print("\n")

# ii. Test the hypothesis and conclude
print("ii. Test the hypothesis and conclude")

# Calculate the Z-statistic
# Z = (sample_mean - population_mean) / (population_std / sqrt(sample_size))
z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Calculate the two-tailed p-value for the Z-statistic
p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

print(f"   Population Mean (μ): {population_mean}")
print(f"   Population Standard Deviation (σ): {population_std}")
print(f"   Sample Size (n): {sample_size}")
print(f"   Sample Mean (x̄): {sample_mean}")
print(f"   Z-statistic: {z_statistic:.4f}")
print(f"   P-value: {p_value:.4f}")

# Conclusion
if p_value < alpha:
    print(f"   Since the p-value ({p_value:.4f}) is less than the significance level ({alpha}), we reject the null hypothesis.")
    print("   Conclusion: There is a statistically significant change in the NYSE P/E value from its previous value.")
else:
    print(f"   Since the p-value ({p_value:.4f}) is greater than or equal to the significance level ({alpha}), we fail to reject the null hypothesis.")
    print("   Conclusion: There is no statistically significant change in the NYSE P/E value from its previous value.")
```

### 3c) The demand for a particular spare part was found to vary from day to day. In a sample study the following information was obtained.

Days	   Mon	   Tue	    Wed	   Thur	  Friday	Saturday
Quantity
demanded	1124	1125	1110	1120	1126	1115

- i. Write the hypothesis.(2 marks)

- ii. Test the hypothesis at 1% level of significance that the number demanded depends upon the day (4 marks)

```python
import numpy as np
from scipy import stats

# Given data
days = ['Mon', 'Tue', 'Wed', 'Thur', 'Friday', 'Saturday']
quantity_demanded = np.array([1124, 1125, 1110, 1120, 1126, 1115])

# Significance level
alpha = 0.01

# i. Write the hypothesis
print("i. Write the hypothesis.")
print("   Null Hypothesis (H0): The demand for the spare part does not depend on the day of the week (i.e., the demand is uniformly distributed across the days).")
print("   Alternative Hypothesis (Ha): The demand for the spare part depends on the day of the week (i.e., the demand is not uniformly distributed across the days).")
print("   Type of test: Chi-squared Goodness-of-Fit test (to compare observed frequencies with expected frequencies under the null hypothesis of uniform distribution).")
print("\n")

# ii. Test the hypothesis at 1% level of significance and conclude
print("ii. Test the hypothesis at 1% level of significance and conclude.")

# Calculate total demand
total_demand = np.sum(quantity_demanded)

# Calculate expected demand per day under the null hypothesis (uniform distribution)
num_days = len(days)
expected_demand = np.full(num_days, total_demand / num_days)

print(f"   Observed Quantities: {quantity_demanded}")
print(f"   Expected Quantities (under H0): {expected_demand}")

# Perform Chi-squared Goodness-of-Fit test
chi2_statistic, p_value = stats.chisquare(f_obs=quantity_demanded, f_exp=expected_demand)

print(f"   Chi-squared Statistic: {chi2_statistic:.4f}")
print(f"   P-value: {p_value:.4f}")

# Conclusion
if p_value < alpha:
    print(f"   Since the p-value ({p_value:.4f}) is less than the significance level ({alpha}), we reject the null hypothesis.")
    print("   Conclusion: There is statistically significant evidence to conclude that the number demanded depends upon the day.")
else:
    print(f"   Since the p-value ({p_value:.4f}) is greater than or equal to the significance level ({alpha}), we fail to reject the null hypothesis.")
    print("   Conclusion: There is no statistically significant evidence to conclude that the number demanded depends upon the day.")
```

#### 3d) General hospitals patient account division has compiled data on the age of accounts receivable. The data collected indicate that the age of the accounts follows a normal distribution with μ = 28 days and σ = 8 days.

- What portion of the accounts is between 20 and 40 days old. (4 mark)

- The hospital administrator is interested in sending reminder letters to the oldest 15% of accounts. How many days old should an account be, before a reminder letter is sent. (2 mark)

```python
import numpy as np
from scipy.stats import norm

# Given parameters for the normal distribution
mu = 28  # Mean age of accounts (days)
sigma = 8  # Standard deviation of account age (days)

print("### Q3d) General hospitals patient account division ###")
print(f"Given: Mean (μ) = {mu} days, Standard Deviation (σ) = {sigma} days\n")

# i. What portion of the accounts is between 20 and 40 days old?
print("i. What portion of the accounts is between 20 and 40 days old?")

# Calculate the Z-scores for 20 and 40 days
z_20 = (20 - mu) / sigma
z_40 = (40 - mu) / sigma

# Calculate the cumulative probabilities for these Z-scores
prob_less_than_20 = norm.cdf(20, loc=mu, scale=sigma)
prob_less_than_40 = norm.cdf(40, loc=mu, scale=sigma)

# The portion between 20 and 40 days is P(X <= 40) - P(X <= 20)
portion_between_20_and_40 = prob_less_than_40 - prob_less_than_20

print(f"   Z-score for 20 days: {z_20:.2f}")
print(f"   Z-score for 40 days: {z_40:.2f}")
print(f"   P(X < 20): {prob_less_than_20:.4f}")
print(f"   P(X < 40): {prob_less_than_40:.4f}")
print(f"   The portion of accounts between 20 and 40 days old is: {portion_between_20_and_40:.4f} or {portion_between_20_and_40*100:.2f}%\n")

# ii. How many days old should an account be, before a reminder letter is sent to the oldest 15% of accounts?
print("ii. How many days old should an account be, before a reminder letter is sent to the oldest 15% of accounts?")

# This means we need to find the value x such that P(X > x) = 0.15
# Or, P(X <= x) = 1 - 0.15 = 0.85

# Use the percent point function (PPF) or inverse CDF to find this value
days_for_reminder = norm.ppf(0.85, loc=mu, scale=sigma)

print(f"   To identify the oldest 15% of accounts, we need to find the value X such that P(X > X_value) = 0.15.")
print(f"   This is equivalent to finding the X_value at the 85th percentile.")
print(f"   An account should be at least {days_for_reminder:.2f} days old before a reminder letter is sent.")
```

### 3e) A pharmaceutical company claims that its new tablet is effective in increasing the height of children. The data of heights (in cm) of 7 children is recorded before and after consuming the tablet.

- Check for normality of the data and perform appropriate test.

- Test the company's claim at a 5% level of significance using the p-value approach.


    ht_before = [121, 125, 130, 120, 145, 126, 134]
    
    ht_after = [130, 129, 148, 122, 147, 130, 148]

```python
import numpy as np
from scipy import stats

# Data of heights (in cm)
ht_before = np.array([121, 125, 130, 120, 145, 126, 134])
ht_after = np.array([130, 129, 148, 122, 147, 130, 148])
alpha = 0.05 # Significance level for the company's claim

print("### Q3e) A pharmaceutical company claims that its new tablet is effective ###\n")

# --- Check for normality of the data ---
print("i. Check for normality of the data and perform appropriate test.")
print("   Hypothesis for Normality Test:")
print("   Null Hypothesis (H0): The data is drawn from a normal distribution.")
print("   Alternative Hypothesis (Ha): The data is not drawn from a normal distribution.")
print("   Test: Shapiro-Wilk test (appropriate for small sample sizes).")
print("   Significance level for normality: We will use alpha = 0.05 for this check.\n")

# Shapiro-Wilk test for ht_before
shapiro_before_stat, shapiro_before_p = stats.shapiro(ht_before)
print(f"   Shapiro-Wilk Test for ht_before:")
print(f"     W-statistic: {shapiro_before_stat:.4f}")
print(f"     P-value: {shapiro_before_p:.4f}")
if shapiro_before_p < alpha:
    print("     Conclusion: Reject H0. ht_before does not appear to be normally distributed.")
else:
    print("     Conclusion: Fail to reject H0. ht_before appears to be normally distributed.")

print("\n")

# Shapiro-Wilk test for ht_after
shapiro_after_stat, shapiro_after_p = stats.shapiro(ht_after)
print(f"   Shapiro-Wilk Test for ht_after:")
print(f"     W-statistic: {shapiro_after_stat:.4f}")
print(f"     P-value: {shapiro_after_p:.4f}")
if shapiro_after_p < alpha:
    print("     Conclusion: Reject H0. ht_after does not appear to be normally distributed.")
else:
    print("     Conclusion: Fail to reject H0. ht_after appears to be normally distributed.")

print("\n--- Assumption for Paired t-test: The differences between paired observations are normally distributed ---\n")
# Check normality of the differences, which is the key assumption for a paired t-test
differences = ht_after - ht_before
shapiro_diff_stat, shapiro_diff_p = stats.shapiro(differences)
print(f"   Shapiro-Wilk Test for Differences (ht_after - ht_before):")
print(f"     W-statistic: {shapiro_diff_stat:.4f}")
print(f"     P-value: {shapiro_diff_p:.4f}")
if shapiro_diff_p < alpha:
    print("     Conclusion: Reject H0. Differences do not appear to be normally distributed. A non-parametric test (e.g., Wilcoxon Signed-Rank test) might be more appropriate in this case, though we will proceed with the t-test as requested.")
else:
    print("     Conclusion: Fail to reject H0. Differences appear to be normally distributed. Paired t-test is appropriate.")

print("\n")

# --- Test the company's claim at a 5% level of significance using the p-value approach ---
print("ii. Test the company's claim at a 5% level of significance using the p-value approach.")
print("   Hypothesis for Paired t-test:")
print("   Null Hypothesis (H0): The tablet is not effective in increasing height (μ_after <= μ_before).")
print("   Alternative Hypothesis (Ha): The tablet is effective in increasing height (μ_after > μ_before).")
print("   This is a one-tailed (right-tailed) paired t-test.")
print(f"   Significance level: alpha = {alpha}\n")

# Perform paired t-test with 'greater' alternative for a one-tailed test
t_statistic, p_value = stats.ttest_rel(ht_after, ht_before, alternative='greater')

print(f"   Paired t-test results:")
print(f"     T-statistic: {t_statistic:.4f}")
print(f"     One-tailed P-value: {p_value:.4f}")

# Conclusion
if p_value < alpha:
    print(f"     Since the one-tailed P-value ({p_value:.4f}) is less than the significance level ({alpha}), we reject the null hypothesis.")
    print("     Conclusion: There is statistically significant evidence to support the company's claim that the tablet is effective in increasing the height of children.")
else:
    print(f"     Since the one-tailed P-value ({p_value:.4f}) is greater than or equal to the significance level ({alpha}), we fail to reject the null hypothesis.")
    print("     Conclusion: There is no statistically significant evidence to support the company's claim that the tablet is effective in increasing the height of children.")
```

### 4b) Consider the insurance.csv file and answer the following questions

- i)	Check whether the BMI data follows normal distribution by using a qq plot and shaipro test ( 3marks)

- ii)	Construct a hypothesis to prove that smoker have higher charges than non smoker and conclude accordingly ( 4 marks)

- iii)	Is there a statisitical difference between insurance charges for male and female and conclude accordingly ( 4 marks)

- iv)	Is the proportion of smokers significantly different in different genders? (4 Marks)


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import statsmodels.api as sm

# Load the dataset (assuming df is already loaded from Q4a, but reloading for self-containment)
try:
    df = pd.read_csv('sample_data/insurance.csv')
    print("Dataset loaded successfully for Q4b.\n")
except FileNotFoundError:
    print("Error: 'insurance.csv' not found. Please ensure the file is in the 'sample_data' directory.\n")
    df = None

if df is not None:
    print("### Q4b) Consider the insurance.csv file ###\n")

    # i) Check whether the BMI data follows normal distribution by using a qq plot and shaipro test
    print("i) Check whether the BMI data follows normal distribution by using a Q-Q plot and Shapiro test.\n")
    alpha_normality = 0.05

    # Q-Q plot for BMI
    plt.figure(figsize=(8, 6))
    sm.qqplot(df['bmi'], line='s')
    plt.title('Q-Q Plot of BMI')
    plt.xlabel('Theoretical Quantiles')
    plt.ylabel('Sample Quantiles')
    plt.grid(True)
    plt.show()

    # Shapiro-Wilk test for BMI
    shapiro_stat_bmi, shapiro_p_bmi = stats.shapiro(df['bmi'])

    print(f"   Shapiro-Wilk Test for BMI:")
    print(f"     W-statistic: {shapiro_stat_bmi:.4f}")
    print(f"     P-value: {shapiro_p_bmi:.4f}")
    if shapiro_p_bmi < alpha_normality:
        print("     Conclusion: Reject H0. BMI does not appear to be normally distributed (p-value < 0.05).")
    else:
        print("     Conclusion: Fail to reject H0. BMI appears to be normally distributed (p-value >= 0.05).")
    print("\nObservation from Q-Q Plot: The points largely follow the straight line, suggesting a normal distribution, but there are some deviations at the tails, indicating slight departures from normality, particularly a heavier right tail. The Shapiro-Wilk test provides a statistical confirmation.\n")
    print("-"*70 + "\n")

    # ii) Construct a hypothesis to prove that smokers have higher charges than non smoker and conclude accordingly
    print("ii) Hypothesis to prove that smokers have higher charges than non-smokers.\n")

    # Separate data for smokers and non-smokers
    smoker_charges = df[df['smoker'] == 'yes']['charges']
    non_smoker_charges = df[df['smoker'] == 'no']['charges']

    print("   Hypothesis for Smoker vs Non-Smoker Charges (assuming equal variances for independent t-test for simplicity, as per previous questions' pattern, although Levene's test could confirm):")
    print("   Null Hypothesis (H0): The mean insurance charges for smokers are less than or equal to non-smokers (μ_smoker <= μ_non_smoker).")
    print("   Alternative Hypothesis (Ha): The mean insurance charges for smokers are higher than non-smokers (μ_smoker > μ_non_smoker).")
    print("   Type of test: One-tailed (right-tailed) independent two-sample t-test.")
    alpha_smoker = 0.05 # Significance level
    print(f"   Significance level: alpha = {alpha_smoker}\n")

    # Perform independent t-test (equal_var=False due to large difference in means and std devs often seen in charges, though question implies equal variances for Q3a, for safety with this dataset we'll use Welch's t-test or state assumption)
    # Given the previous questions assumed equal variances, we'll proceed with equal_var=True for consistency, but note the potential for unequal variances.
    t_stat_smoker, p_val_smoker = stats.ttest_ind(smoker_charges, non_smoker_charges, equal_var=True, alternative='greater')

    print(f"   Smoker Charges Mean: {smoker_charges.mean():.2f}")
    print(f"   Non-Smoker Charges Mean: {non_smoker_charges.mean():.2f}")
    print(f"   T-statistic: {t_stat_smoker:.4f}")
    print(f"   One-tailed P-value: {p_val_smoker:.4f}")

    if p_val_smoker < alpha_smoker:
        print(f"   Since the one-tailed P-value ({p_val_smoker:.4f}) is less than the significance level ({alpha_smoker}), we reject the null hypothesis.")
        print("   Conclusion: There is statistically significant evidence to prove that smokers have higher insurance charges than non-smokers.")
    else:
        print(f"   Since the one-tailed P-value ({p_val_smoker:.4f}) is greater than or equal to the significance level ({alpha_smoker}), we fail to reject the null hypothesis.")
        print("   Conclusion: There is no statistically significant evidence to prove that smokers have higher insurance charges than non-smokers.")
    print("\n" + "-"*70 + "\n")

    # iii) Is there a statisitical difference between insurance charges for male and female and conclude accordingly
    print("iii) Statistical difference between insurance charges for male and female.\n")

    # Separate data for male and female charges
    male_charges = df[df['sex'] == 'male']['charges']
    female_charges = df[df['sex'] == 'female']['charges']

    print("   Hypothesis for Male vs Female Charges (assuming equal variances):")
    print("   Null Hypothesis (H0): There is no difference in mean insurance charges between males and females (μ_male = μ_female).")
    print("   Alternative Hypothesis (Ha): There is a difference in mean insurance charges between males and females (μ_male ≠ μ_female).")
    print("   Type of test: Two-tailed independent two-sample t-test.")
    alpha_gender = 0.05 # Significance level
    print(f"   Significance level: alpha = {alpha_gender}\n")

    # Perform independent t-test (two-tailed)
    t_stat_gender, p_val_gender = stats.ttest_ind(male_charges, female_charges, equal_var=True)

    print(f"   Male Charges Mean: {male_charges.mean():.2f}")
    print(f"   Female Charges Mean: {female_charges.mean():.2f}")
    print(f"   T-statistic: {t_stat_gender:.4f}")
    print(f"   Two-tailed P-value: {p_val_gender:.4f}")

    if p_val_gender < alpha_gender:
        print(f"   Since the two-tailed P-value ({p_val_gender:.4f}) is less than the significance level ({alpha_gender}), we reject the null hypothesis.")
        print("   Conclusion: There is a statistically significant difference between insurance charges for males and females.")
    else:
        print(f"   Since the two-tailed P-value ({p_val_gender:.4f}) is greater than or equal to the significance level ({alpha_gender}), we fail to reject the null hypothesis.")
        print("   Conclusion: There is no statistically significant difference between insurance charges for males and females.")
    print("\n" + "-"*70 + "\n")

    # iv) Is the proportion of smokers significantly different in different genders?
    print("iv) Is the proportion of smokers significantly different in different genders?\n")

    # Create a contingency table
    contingency_table = pd.crosstab(df['sex'], df['smoker'])
    print("   Contingency Table (Gender vs Smoker Status):")
    print(contingency_table)

    print("\n   Hypothesis for Proportion of Smokers in different genders:")
    print("   Null Hypothesis (H0): The proportion of smokers is independent of gender.")
    print("   Alternative Hypothesis (Ha): The proportion of smokers is dependent on gender.")
    print("   Type of test: Chi-squared test for independence.")
    alpha_prop = 0.05 # Significance level
    print(f"   Significance level: alpha = {alpha_prop}\n")

    # Perform Chi-squared test for independence
    chi2_stat_prop, p_val_prop, dof_prop, ex_freq_prop = stats.chi2_contingency(contingency_table)

    print(f"   Chi-squared Statistic: {chi2_stat_prop:.4f}")
    print(f"   P-value: {p_val_prop:.4f}")
    print(f"   Degrees of Freedom: {dof_prop}")

    if p_val_prop < alpha_prop:
        print(f"   Since the P-value ({p_val_prop:.4f}) is less than the significance level ({alpha_prop}), we reject the null hypothesis.")
        print("   Conclusion: There is a statistically significant difference in the proportion of smokers between different genders.")
    else:
        print(f"   Since the P-value ({p_val_prop:.4f}) is greater than or equal to the significance level ({alpha_prop}), we fail to reject the null hypothesis.")
        print("   Conclusion: There is no statistically significant difference in the proportion of smokers between different genders.")
```
